# [Supertrend Indicator](https://www.investopedia.com/supertrend-indicator-7976167)


## Summary: 
* The Supertrend Indicator can help identify trends, manage risk, and confirm market tendencies.
* The indicator is limited by its lagging nature, is not very flexible, and can send up false signals.

## Buy/Sell signals: 
* Buy signal: The super trend line drops below the closing price and changes to green.
* Sell signal: The super trend line closes above the closing price and turns red.

```python
import pandas_ta as ta

df.ta.supertrend(append=True)
"""
Returns:
    pd.DataFrame: SUPERT (trend), SUPERTd (direction), SUPERTl (long), SUPERTs (short) columns.
"""

df.tail()
```
