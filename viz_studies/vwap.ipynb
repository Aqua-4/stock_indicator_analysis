# [VWAP Indicator](https://school.stockcharts.com/doku.php?id=technical_indicators:vwap_intraday)


## Investors use the VWAP to: 
* Identify trends
* Find potential price reversal points
* Determine the average price at which a stock is traded

## VWAP signals: 
* Bearish market: When the price is below the VWAP
* Bullish market: When the price is above the VWAP
* Buying demand: When the VWAP is rising
* Selling: When the VWAP is falling
* Potential reversal in sentiment: When the VWAP crosses above or below

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from tqdm import tqdm

_df = df[df['action'] != 'hold']
# _df=_df.head(100)

# Create the multiline chart using Plotly
fig = go.Figure()



# Create the multiline chart using Plotly
fig = go.Figure()


# Add traces for high, low, BBU, and BBL
fig.add_trace(go.Scatter( x=_df.index, y=_df['open'], mode='lines', name='Open'))
fig.add_trace(go.Scatter( x=_df.index, y=_df['high'], mode='lines', name='High'))
fig.add_trace(go.Scatter( x=_df.index, y=_df['low'], mode='lines', name='Low'))
fig.add_trace(go.Scatter( x=_df.index, y=_df['close'], mode='lines', name='Close'))
fig.add_trace(go.Scatter(x=_df.index, y=_df['VWAP_D'], mode='lines+lines', name='VWAP'))

# TODO: this thing does not need for loop
# Add markers for 'buy' and 'sell' actions
for index, row in tqdm(_df.iterrows(), desc='Marking actions...'):

    if row['action'] == 'strong_buy':
        fig.add_trace(go.Scatter(
            x=[index],
            y=[row['low']],
            mode='markers',
            marker=dict(color='green', symbol='star-triangle-up'),
            name='Buy',
            showlegend=False
        ))
    elif row['action'] == 'strong_sell':
        fig.add_trace(go.Scatter(
            x=[index],
            y=[row['high']],
            mode='markers',
            marker=dict(color='red', symbol='star-triangle-down'),
            name='Sell',
            showlegend=False
        ))


# Update layout
fig.update_layout(
    xaxis=dict(title='Date'),
    yaxis=dict(title='Value'),
    title='Multiline Chart with Datetime X-axis',
    showlegend=True,
    legend=dict(x=0, y=1),
    xaxis_tickangle=-45
)


# Show the Plotly figure
fig.show()
```
